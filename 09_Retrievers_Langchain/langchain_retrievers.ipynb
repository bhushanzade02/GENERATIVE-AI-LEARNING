```python
#2️⃣ Import required modules
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
```

```python
#3️⃣ Create sample documents
documents = [
    Document(page_content="LangChain helps build LLM applications"),
    Document(page_content="Vector databases store embeddings"),
    Document(page_content="Chroma is used for similarity search"),
    Document(page_content="Hugging Face provides embedding models")
]
```

```python
# 4️⃣ Load Hugging Face embedding model

# 👉 This does NOT need an API key (runs locally in Colab)
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)


```

```python
#5️⃣ Create Vector Store (Chroma)
vectorstore = Chroma.from_documents(
    documents=documents,
    embedding=embedding_model
)
```

```python
#6️⃣ Create Vector Retriever

retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 2}
)
```

```python
# 7️⃣ Query the retriever



query = "What is chroma used for?"
# query = "What is a vector database?"

results = retriever.invoke(query)

for i, doc in enumerate(results):
    print(f"Result {i+1}: {doc.page_content}")
```

```python
for i, doc in enumerate(results):
    print(f"\n--- Result {i+1} ---")
    print(doc.page_content)
```

# **03 MMR**

```python
#2️⃣ Import required modules
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_core.documents import Document
from langchain_community.vectorstores import FAISS

# Sample documents
docs = [
    Document(page_content="LangChain makes it easy to work with LLMs."),
    Document(page_content="LangChain is used to build LLM based applications."),
    Document(page_content="Chroma is used to store and search document embeddings."),
    Document(page_content="Embeddings are vector representations of text."),
    Document(page_content="MMR helps you get diverse results when doing similarity search."),
    Document(page_content="LangChain supports Chroma, FAISS, Pinecone, and more."),
]



# Load Hugging Face embedding model

# This does NOT need an API key (runs locally in Colab)
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)



#  Create the FAISS vector store from documents
vectorstore = FAISS.from_documents(
    documents=docs,
    embedding=embedding_model
)


# Enable MMR in the retriever
retriever = vectorstore.as_retriever(
    search_type="mmr",                   # <-- This enables MMR
    search_kwargs={"k": 3, "lambda_mult": 0.5}  # k = top results, lambda_mult = relevance-diversity balance
)



query = "What is langchain?"
results = retriever.invoke(query)



for i, doc in enumerate(results):
    print(f"\n--- Result {i+1} ---")
    print(doc.page_content)


```
