```python
from langchain_community.retrievers import WikipediaRetriever
# Initialize the retriever (optional: set language and top_k)
retriever = WikipediaRetriever(top_k_results=2, lang="en")
# Define your query
query = "the geopolitical history of india and pakistan from the perspective of a chinese"

# Get relevant Wikipedia documents
docs = retriever.invoke(query)
# Print retrieved content
for i, doc in enumerate(docs):
    print(f"\n--- Result {i+1} ---")
    print(f"Content:\n{doc.page_content}...")  # truncate for display
```

```python
# 2️⃣ Import required modules
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
```

```python
#3️⃣ Create sample documents
documents = [
    Document(page_content="LangChain helps build LLM applications"),
    Document(page_content="Vector databases store embeddings"),
    Document(page_content="Chroma is used for similarity search"),
    Document(page_content="Hugging Face provides embedding models")
]
```

```python
# 4️⃣ Load Hugging Face embedding model

# 👉 This does NOT need an API key (runs locally in Colab)
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)


```

```python
#5️⃣ Create Vector Store (Chroma)
vectorstore = Chroma.from_documents(
    documents=documents,
    embedding=embedding_model
)
```

```python
#6️⃣ Create Vector Retriever

retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 2}
)
```

```python
# 7️⃣ Query the retriever



query = "What is chroma used for?"
# query = "What is a vector database?"

results = retriever.invoke(query)

for i, doc in enumerate(results):
    print(f"Result {i+1}: {doc.page_content}")
```
